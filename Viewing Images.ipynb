## Viewing Images in Python

2 primary ways for quick display of images
- pyplot and opencv
- skimage can also be used.. io.imshow

### Using skimage

io.imshow(image)

```python
from skimage import io
```

```python
img = io.imread(r"C:\Users\Owner\python\Datasets\Training\Fire\a.jpg")   #reading image
io.imshow(img)         # displaying image
```

### Using Matplotlib

plt.imshow(image)

```python
import matplotlib.pyplot as plt

plt.imshow(img)
```

```python
#Colormaps...  
plt.imshow(img, cmap="hot")
#Not going to do anything as the input image is RGB
```

```python
# Convert it to grayscale image

gray_img = io.imread(r"C:\Users\Owner\python\Datasets\Training\Fire\a.jpg", as_gray=True)
plt.imshow(gray_img,cmap="hot")
```

```python
plt.imshow(gray_img, cmap="Blues")
```

```python
# Multiple plots using pyplot

fig = plt.figure(figsize=(10,10))

ax1 = fig.add_subplot(2,2,1)
ax1.imshow(gray_img, cmap='hot')
ax1.title.set_text('1st')

ax2 = fig.add_subplot(2,2,2)
ax2.imshow(gray_img, cmap='Blues')
ax2.title.set_text('2nd')

ax3 = fig.add_subplot(2,2,3)
ax3.imshow(gray_img, cmap='gray')
ax3.title.set_text('3rd')

ax4 = fig.add_subplot(2,2,4)
ax4.imshow(gray_img, cmap='nipy_spectral')
ax4.title.set_text('4th')

plt.show()
```

### Using openCV

cv2.imshow()

```python
import cv2
```

```python
gray_img = cv2.imread(r"C:\Users\Owner\python\Datasets\Training\Fire\a.jpg", 0)
color_img = cv2.imread(r"C:\Users\Owner\python\Datasets\Training\Fire\a.jpg", 1)
```

- Use the function cv2.imshow() to display an image in a window. 
- First argument is the window name which is a string. second argument is our image. 

```python
cv2.imshow('pic from skimage import', img)  #Shows weird colors as R and B channels are swapped
cv2.imshow('color pic from opencv', color_img)
cv2.imshow('gray image from opencv', gray_img)

# Maintain output window until 
# user presses a key or 1000 ms (1s)
cv2.waitKey(0)          

#destroys all windows created
cv2.destroyAllWindows() 
```
